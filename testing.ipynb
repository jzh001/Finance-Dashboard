```python
import numpy as np
import pandas as pd
import finance_data_scraper as scraper
import dashboard as dash
import altair as alt
```

```python
data = scraper.getYahooHistorical("AAPL", "1y")
data
```

```python
def getResistSupportChart(df):
    x = np.arange(len(df))
    y = df['close'].values

    regression = np.polyfit(x, y, deg=1)
    print(regression)
    resistance_slope = regression[0]
    resistance_intercept = regression[1]
    support_slope = -resistance_slope
    support_intercept = resistance_intercept
    base_chart = alt.Chart(df.reset_index()).encode(
        x="date:T",
        y="close:Q",
    ).properties(height=350)

    price_line = base_chart.mark_line().encode(
        tooltip=["date:T", "close:Q"],
    )

    resistance_line = base_chart.transform_regression('x', 'y', method='linear', params=True).mark_line(color='red').encode(
        x='date:T',
        y=alt.Y('resistance_y', axis=alt.Axis(title='Price')),
    ).transform_calculate(
        resistance_y=f"x * {resistance_slope} + {resistance_intercept}"
    )

    support_line = base_chart.transform_regression('x', 'y', method='linear', params=True).mark_line(color='green').encode(
        x='date:T',
        y=alt.Y('support_y', axis=alt.Axis(title='Price')),
    ).transform_calculate(
        support_y=f"x * {support_slope} + {support_intercept}"
    )

    chart = (price_line + resistance_line + support_line).interactive()
    return chart
```

```python
getResistSupportChart(data)
```
